```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
r = 1000
f = 100
t = 0

rabbit_birth_rate = 0.7
hunting_parameter = 0.0007
fox_death_rate = 0.5
hunting_foxes_birth_rate = 0.0007
h = 0.001

R = []
F = []
T = []
```

```python
while t <= 100:
    R.append(r)
    F.append(f)
    T.append(t)
    
    r_old = r
    f_old = f
    r += (rabbit_birth_rate * r - hunting_parameter * f * r) * h
    f += (hunting_foxes_birth_rate * r_old * f - fox_death_rate * f) * h
    t += h
```

```python
plt.figure(figsize=[15, 6])
plt.plot(T, R, label="Кролики")
plt.plot(T, F, label="Лисы")
plt.grid()
plt.legend()
plt.title("Модель Хищник-Жертва")
plt.xlabel("Время")
plt.ylabel("Популяция")
plt.show()
```
