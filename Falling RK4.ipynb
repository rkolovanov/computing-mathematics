```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def derivative_vx(vx, vy):
    return -k * vx * math.sqrt(vx ** 2 + vy ** 2)

def derivative_vy(vx, vy):
    return -k * vy * math.sqrt(vx ** 2 + vy ** 2) - g

def derivative_x(vx):
    return vx

def derivative_y(vy):
    return vy
```

```python
t0 = 0
x0 = 0
y0 = 50
v0 = 30
angle = math.radians(45)
gamma = 0.25
g = 9.8
m = 3
k = gamma / m
h = 0.01

t = t0
x = x0
y = y0
vx = v0 * math.cos(angle)
vy = v0 * math.sin(angle)

X = []
Y = []
VX = []
VY = []
T = []
```

```python
while y >= 0:
    X.append(x)
    Y.append(y)
    VX.append(vx)
    VY.append(vy)
    T.append(t)
    
    k1_x = derivative_x(vx)
    k1_y = derivative_y(vy)
    k1_vx = derivative_vx(vx, vy)
    k1_vy = derivative_vy(vx, vy)
    
    k2_x = derivative_x(vx + h * k1_vx / 2)
    k2_y = derivative_y(vy + h * k1_vy / 2)
    k2_vx = derivative_vx(vx + h * k1_vx / 2, vy + h * k1_vy / 2)
    k2_vy = derivative_vy(vx + h * k1_vx / 2, vy + h * k1_vy / 2)
    
    k3_x = derivative_x(vx + h * k2_vx / 2)
    k3_y = derivative_y(vy + h * k2_vy / 2)
    k3_vx = derivative_vx(vx + h * k2_vx / 2, vy + h * k2_vy / 2)
    k3_vy = derivative_vy(vx + h * k2_vx / 2, vy + h * k2_vy / 2)
    
    k4_x = derivative_x(vx + h * k3_vx)
    k4_y = derivative_y(vy + h * k3_vy)
    k4_vx = derivative_vx(vx + h * k3_vx, vy + h * k3_vy)
    k4_vy = derivative_vy(vx + h * k3_vx, vy + h * k3_vy)
    
    x += h * (k1_x + 2 * k2_x + 2 * k3_x + k4_x) / 6
    y += h * (k1_y + 2 * k2_y + 2 * k3_y + k4_y) / 6
    vx += h * (k1_vx + 2 * k2_vx + 2 * k3_vx + k4_vx) / 6
    vy += h * (k1_vy + 2 * k2_vy + 2 * k3_vy + k4_vy) / 6
    t += h
```

```python
plt.figure(figsize=[10, 10])
plt.xlim(-1, 75)
plt.ylim(-1, 75)
plt.plot(X, Y)
plt.grid()
plt.xlabel("Координата X")
plt.ylabel("Координата Y")
plt.show()

plt.plot(T, VX)
plt.grid()
plt.xlabel("Время, сек.")
plt.ylabel("Скорость по X")
plt.show()

plt.plot(T, VY)
plt.grid()
plt.xlabel("Время, сек.")
plt.ylabel("Скорость по Y")
plt.show()
```
