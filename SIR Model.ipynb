```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 100000
i = 1
r = 0
s = N - i - r
beta = 0.5
mu = 0.1
h = 0.001
t = 0

S = []
I = []
R = []
T = []
```

```python
while t <= 80:
    S.append(s)
    I.append(i)
    R.append(r)
    T.append(t)
    
    s_old = s
    i_old = i
    s = s - beta * s * i * h / N
    i = i + (beta * s_old * i / N - mu * i_old) * h
    r = r + mu * i_old * h
    t += h
```

```python
plt.figure(figsize=[15, 6])
plt.plot(T, S, label="Не болевшие")
plt.plot(T, I, label="Инфицированные")
plt.plot(T, R, label="Переболевшие")
plt.grid()
plt.legend()
plt.title("SIR Модель")
plt.xlabel("Время")
plt.ylabel("Популяция")
plt.show()
```
