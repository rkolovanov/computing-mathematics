```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
k = 0.8
x = 10
v = 0
a = -k * x
h = 0.001
t = 0

A = []
V = []
X = []
T = []
```

```python
while t <= 50:
    V.append(v)
    X.append(x)
    T.append(t)
    A.append(a)
    
    a = -k * x
    v = v + h * a
    x = x + h * v
    t += h
```

```python
plt.figure(figsize=[15, 6])
plt.plot(T, X, label=" X")
plt.grid()
plt.legend()
plt.xlabel("t, ")
plt.ylabel("x(t), ")
plt.show()

plt.figure(figsize=[15, 6])
plt.plot(T, V, label="  X")
plt.grid()
plt.legend()
plt.xlabel("t, ")
plt.ylabel("v(t), /")
plt.show()

plt.figure(figsize=[15, 6])
plt.plot(T, A, label="  X")
plt.grid()
plt.legend()
plt.xlabel("t, ")
plt.ylabel("a(t), /^2")
plt.show()
```
