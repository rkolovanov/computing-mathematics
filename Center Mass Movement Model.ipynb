```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
m = 1
x = 0
v = 0
h = 0.001
t = 0

def F(t):
    return m

X = []
V = []
T = []
```

```python
while t <= 10:
    X.append(x)
    V.append(v)
    T.append(t)
    
    v = v + h * F(t) / m
    x = x + h * v
    t += h
```

```python
plt.figure(figsize=[15, 6])
plt.plot(T, X, label="Координата X")
plt.plot(T, V, label="Скорость")
plt.grid()
plt.legend()
plt.xlabel("t, с")
plt.show()
```
