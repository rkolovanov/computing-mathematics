```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def derivative_x1(x1, x2):
    return x1 - x1 ** 3 - x2

def derivative_x2(x1):
    return x1
    
```

```python
x1 = 1
x2 = 2
h = 0.001
t = 0

X1 = []
X2 = []
T = []
```

```python
while t <= 50:
    X1.append(x1)
    X2.append(x2)
    T.append(t)
    
    x1_old = x1
    x2_old = x2
    
    k1_x1 = derivative_x1(x1_old, x2_old)
    k1_x2 = derivative_x2(x1_old)
    
    k2_x1 = derivative_x1(x1_old + h * k1_x1 / 2, x2_old + h * k1_x2 / 2)
    k2_x2 = derivative_x2(x1_old + h * k1_x1 / 2)
    
    k3_x1 = derivative_x1(x1_old + h * k2_x1 / 2, x2_old + h * k2_x2 / 2)
    k3_x2 = derivative_x2(x1_old + h * k2_x1 / 2)
    
    k4_x1 = derivative_x1(x1_old + h * k3_x1, x2_old + h * k3_x2)
    k4_x2 = derivative_x2(x1_old + h * k3_x1)
             
    x1 += h * (k1_x1 + 2 * k2_x1 + 2 * k3_x1 + k4_x1) / 6
    x2 += h * (k1_x2 + 2 * k2_x2 + 2 * k3_x2 + k4_x2) / 6
    t += h
```

```python
plt.figure(figsize=[15, 6])
plt.plot(T, X1, label="X1")
plt.plot(T, X2, label="X2")
plt.grid()
plt.legend()
plt.title("Модель электрического осциллятора")
plt.xlabel("Время")
plt.ylabel("Значение X")
plt.show()
```
