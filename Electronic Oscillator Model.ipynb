```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x1 = 1
x2 = 2
h = 0.001
t = 0

X1 = []
X2 = []
T = []
```

```python
while t <= 50:
    X1.append(x1)
    X2.append(x2)
    T.append(t)
    
    x1_old = x1
    x1 = x1 + (x1 - x1 ** 3 - x2) * h
    x2 = x2 + x1_old * h
    t += h
```

```python
plt.figure(figsize=[15, 6])
plt.plot(T, X1, label="X1")
plt.plot(T, X2, label="X2")
plt.grid()
plt.legend()
plt.title("Модель электрического осциллятора")
plt.xlabel("Время")
plt.ylabel("Значение X")
plt.show()
```
